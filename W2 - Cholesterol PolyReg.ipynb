```dg
plt = import '/matplotlib/pyplot'
pd = import '/pandas'
np = import '/numpy'
PolynomialFeatures = import '/sklearn/preprocessing/PolynomialFeatures'
LinearRegression = import '/sklearn/linear_model/LinearRegression'
train_test_split = import '/sklearn/model_selection/train_test_split' 
```

```dg
veriseti = pd.read_table "https://gist.githubusercontent.com/LeaveNhA/6eaea566d9faf477205307446f43c3ff/raw/46fc24a48ac2c85f9c6f8e0858a1795fdf598d2c/hasta.data" sep:" "
X = np.array <| list <| map list <| map (bind take 3) veriseti.iloc
```

```dg
y = veriseti !! 'KOLESTROL'
y.shape
```

```dg
x = pd.DataFrame (np.array <| list <| map list' (veriseti !! 'KILO') (veriseti !! 'YAS')) columns:['KILO', 'YAS']
Y = y
```

```dg
[x_eğitim, x_test, Y_eğitim, Y_test] = train_test_split x Y test_size:0.3 random_state:5
```

```dg
#RNM{derece:degree}
#RNM{polinom_özellikleri:polinom_özellikler}
derece = 2
polinom_özellikleri = PolynomialFeatures degree:derece
x_eğitim_polinom = polinom_özellikleri.fit_transform x_eğitim
```

```dg
#NEW{Görünüşe göre SKLearn, get_feature_names işlevini `önerilmez` olarak işaretlemekte haklıymış
#    Şu anda, sütun isimleri gelmiyor. Sanırım sürüm farkı. ¯\_(ツ)_/¯ }

print polinom_özellikleri.get_feature_names!
```

```dg
model = LinearRegression!
model.fit x_eğitim_polinom Y_eğitim
```

```dg
x_test_polinom = polinom_özellikleri.fit_transform x_test
"R-kare: {:.4f}".format $ model.score x_test_polinom Y_test
```

```dg
print model.intercept_
print model.coef_
```

```dg
"Kolestrol= {:.3f} - {:.3f} * Kilo + {:.3f} * Yaş + {:.3f} * Kilo² - {:.3f} * (Kilo - Yaş) + {:.3f} * Yaş²".format model.intercept_ *: (tail model.coef_)
```

```dg
#RNM{derece:degree}
#RNM{polinom_özellikleri_3d:polynominal_features}
#RNM{x_polinom:x_poly}
#EaV{z_surf: deterministik olan z işlevinin bir hesaplaması olarak açılmıştır, z(x_surf, y_surf)}

# 3D Hiperdüzlem Grafiği:
fig = plt.figure figsize:(18, 15)
ax = fig.add_subplot 111 projection:'3d'

ax.scatter (x !! 'KILO') (x !! 'YAS') Y c:'b'
ax.set_xlabel "Kilo"
ax.set_ylabel "Yaş"
ax.set_zlabel "Kolestrol"

# Kilo için:
x_surf = np.arange 0 120 5
# Yaş için:
y_surf = np.arange 0 80 5
[x_surf, y_surf] = np.meshgrid x_surf y_surf

# 2. Dereceden Polinom Regresyonu Eğitimi:
derece = 2
polinom_özellikleri_3d = PolynomialFeatures degree:derece
x_polinom = polinom_özellikleri_3d.fit_transform x

print "Öznitelikler:" $ polinom_özellikleri.get_feature_names!
    
# Doğrusal Regresyon Eğitimi:
model = LinearRegression!
model.fit x_polinom Y

# Eğitilen Modele Dayalı z(Kolestrol) Hesaplama:
# Referans için matematiksel modelin açıklaması, In[20]'ye danışabilirsiniz.
z = x_ y_ -> (map (*) (tail model.coef_) [x_, y_, x_ ** 2, x_ * y_, y_ ** 2]) |> sum |> (+ model.intercept_)

z_surf = z x_surf y_surf
ax.plot_surface x_surf y_surf z_surf rstride: 1 cstride: 1 color: 'NONE' alpha: 0.2

                
fig
```
