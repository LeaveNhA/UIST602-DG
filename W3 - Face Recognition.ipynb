```dg
except
  module_err_ =>
    SVC = import '/sklearn/svm/SVC'
    RandomizedPCA = import '/sklearn/decomposition/PCA'
    make_pipeline = import '/sklearn/pipeline/make_pipeline'
  module_err_ :: ModuleNotFoundError =>
    print 'Modül(ler) yüklü değil ya da bulunamıyor.'
```

```dg
pca = RandomizedPCA n_components: 150
                    whiten: True 
                    random_state: 42
svc = SVC kernel: 'rbf' class_weight: 'balanced'
model = make_pipeline pca svc
```

```dg
except
  module_err_ =>
    model_selection = import '/sklearn/model_selection'
    [train_test_split, GridSearchCV] =
      (map (nm -> (getattr model_selection nm None)) ['train_test_split', 'GridSearchCV']) |> list
  module_err_ :: ModuleNotFoundError =>
    print 'Modül(ler) yüklü değil ya da bulunamıyor.'
```

```dg
[X_eğitim, X_test, y_eğitim, y_test] = train_test_split yüzler.data
                                                        yüzler.target
                                                        random_state: 42
```

```dg
#QST{Neden CPU zamanı işleci gösteriliyor?
#    Kayan noktalı hesaplamada Grafik kartı daha verimli.
#    Ekran kartımın kullanılması için ne yapabilirim, araştır!}
#TDO{Maalesef, şu anda bu kodu yazarken, Idg çekirdeği %time sihirli
#    ifadesini desteklemiyor. Bu sebeple ifade değiştirildi.}

param_grid = {('svc__C', [1, 5, 10, 50]),
              ('svc__gamma', [0.0001, 0.0005, 0.001, 0.005])}

grid = GridSearchCV model param_grid

# Operasyon süresi:
grid.fit X_eğitim y_eğitim
print grid.best_params_
```

```dg
#RNM{model:model_, örnek özelinde, önceki değerin ezilmemesi için.}

model_ = grid.best_estimator_
yfit = model_.predict X_test
```

```dg
fig, ax = plt.subplots 4 6 figsize: (10, 8)

for [i, ax_] in enumerate ax.flat =>
    ax_.imshow ((X_test !! i) |>.reshape 62 47) cmap: 'bone'
    ax_.set xticks: [] yticks: []
    ax_.set_ylabel (((yüzler.target_names !! (yfit !! i)) |>.split!) !! -1)
                   color: (if
                            (yfit !! i) == (y_test !! i) => 'blue'
                            otherwise => 'red')
    
fig
```
