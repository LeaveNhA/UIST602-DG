```dg
#NEW {numpy.c_, sadece basit bir fermuarlama (zip) operasyonu gibi duruyor. }

x = pd.DataFrame (zip (veriseti !! 'KILO') (veriseti !! 'YAS')) columns: ['Kilo', 'Yaş']
Y = veriseti !! 'KOLESTROL'

[x_train, x_test, Y_train, Y_test] = train_test_split x Y test_size:0.3 random_state:5
```

```dg
# Gözlem için verilerin yapılarına bakıyoruz:
print $ "x_train.shape: {}.{n}Y_train: {}.".format x_train.shape Y_train.shape n:"\n"

print $ "x_test.shape: {}.{n}Y_test: {}.".format x_test.shape Y_test.shape n:"\n"
```

```dg
# Doğrusal regresyon için bağımlılık:
LinearRegression = import '/sklearn/linear_model/LinearRegression'
```

```dg
model = LinearRegression!
model.fit x_train Y_train
```

```dg
kolestrol_tahmin = model.predict x_test
```

```dg
"R-kare: {:.4f}".format $ model.score x_test Y_test
```

```dg
# Metrikler için bağımlılık:
mean_squared_error = import '/sklearn/metrics/mean_squared_error'

mse = mean_squared_error Y_test kolestrol_tahmin
print mse
```

```dg
# Bir de grafikle görelim:
figür = plt.figure!

plt.scatter Y_test kolestrol_tahmin
plt.xlabel "Gerçek Kolestrol"
plt.ylabel "Tahmini Kolestrol"
plt.title "Gerçek Kolestrol v Tahmin Edilen Kolestrol"

figür
```

```dg
# Model katsayıları:
print model.intercept_
print model.coef_
```

```dg
"Böylece model; `Kolestrol = {:.3f} + {:.3f} * Kilo + {:.3f} * Yaş` olarak ifade edilir.".format model.intercept_ (model.coef_ !! 0) (model.coef_ !! 1)
```
